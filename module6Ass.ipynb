```python

Todo_file="todo.txt"

def add_task():
  task=input("Enter the task:")
  if task:
    print("\nYour Tasks:")
    with open(Todo_file,"a") as file:
      file.write(task+"\n")
      print("Task added successfully!")
  else:
    print("Empty task not added.")

def view_tasks():
  try:
        with open(Todo_file, "r") as file:
            tasks = file.readlines()
            if tasks:
                print("\nYour Tasks:")
                for i, task in enumerate(tasks, 1):
                    print(f"{i}. {task.strip()}")
            else:
                print("\nNo tasks found.")
  except FileNotFoundError:
        print("\nNo tasks found.")

def clear_tasks():
  with open(Todo_file,"w") as file:
    file.write("")
  print("All tasks cleared.")


def main():
    while True:
        print("\n--- To-Do List Menu ---")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Clear All Tasks")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ").strip()
        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            clear_tasks()
        elif choice == "4":
            print("Exiting... Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

main()
```

```python
Test_file="test.txt"

def add_sentence():
  sen=input("Enter the sentence:")
  if sen:
    print("\nYour Sentence:")
    with open(Test_file,"a") as file:
      file.write(sen+"\n")
      print("Sentence added successfully!")
  else:
    print("Empty sentence not added.")

def search():
    search_word=input("Enter the word to search:").strip().lower()
    if search_word:
        count=0
        try:
            with open(Test_file,"r") as file:
                for i, line in enumerate(file, 1):
                    if search_word in line.lower():
                        count+=1
                        print(f"Found '{search_word}' on line {i}: {line.strip()}")
            if count == 0:
                print(f"'{search_word}' not found in {Test_file}")
            else:
                print(f"Found '{search_word}' {count} time(s).")
        except FileNotFoundError:
            print(f"{Test_file} not found.")


def main():
    while True:
        print("1. Add Sentence")
        print("2. Search sentence")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ").strip()
        if choice == "1":
            add_sentence()
        elif choice == "2":
            search()
        elif choice == "3":
            print("Exiting... Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

main()
```

```python
def copy_file():
    source = input("Enter the source file name: ")
    destination = input("Enter the destination file name: ")

    try:
        with open(source, 'r') as src_file:
            content = src_file.read()

        with open(destination, 'w') as dest_file:
            dest_file.write(content)

        print(f"Content copied from '{source}' to '{destination}' successfully.")

    except FileNotFoundError:
        print(f"Error: The file '{source}' does not exist.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

copy_file()
```

```python
with open("config.txt", "w") as file:
    file.write("host=localhost\n")
    file.write("port=5432\n")
    file.write("username=admin\n")

def read_config(file_name):
    config = {}

    try:
        with open(file_name, 'r') as file:
            for line in file:
                line = line.strip()
                if line and not line.startswith('#'):
                    if '=' in line:
                        key, value = line.split('=', 1)
                        config[key.strip()] = value.strip()
        return config

    except FileNotFoundError:
        print(f"Error: Configuration file '{file_name}' not found.")
        return None

    except Exception as e:
        print(f"An error occurred: {e}")
        return None


def main():
    file_name = input("Enter the configuration file name: ")
    config = read_config(file_name)

    if config is None:
        return

    print("\nConfiguration Loaded:")
    for key, value in config.items():
        print(f"{key} = {value}")

    while True:
        query = input("\nEnter a key to retrieve its value (or type 'exit' to quit): ").strip()
        if query.lower() == 'exit':
            print("Exiting.")
            break
        elif query in config:
            print(f"{query} = {config[query]}")
        else:
            print(f"Key '{query}' not found in configuration.")


main()
```

```python
Test_file="test.txt"

def add_sentence():
  sen=input("Enter the sentence:")
  if sen:
    print("\nYour Sentence:")
    with open(Test_file,"a") as file:
      file.write(sen+"\n")
      print("Sentence added successfully!")
  else:
    print("Empty sentence not added.")

def count():
    word_count = 0
    line_count = 0
    try:
        with open(Test_file,"r") as file:
            for line in file:
                line_count += 1
                words = line.split()
                word_count += len(words)
        print(f"\nFile Statistics for '{Test_file}':")
        print(f"Total lines: {line_count}")
        print(f"Total words: {word_count}")
    except FileNotFoundError:
        print(f"\nError: The file '{Test_file}' does not exist.")
    except Exception as e:
        print(f"\nAn error occurred: {e}")



def main():
    while True:
        print("1. Add Sentence")
        print("2. count word and lines")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ").strip()
        if choice == "1":
            add_sentence()
        elif choice == "2":
            count()
        elif choice == "3":
            print("Exiting... Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

main()
```
