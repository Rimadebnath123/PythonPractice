```python
import pandas as pd
```

```python
# Set display option to avoid line wrapping
pd.set_option('display.max_columns', None)  # show all columns
pd.set_option('display.expand_frame_repr', False)  # prevent wrapping to multiple lines
df = pd.read_csv("employee.csv")
print(df)
# print(df.head())
```

```python
# Group by Department and calculate average salary
avg_salary_by_dept = df.groupby("Department")["Salary"].mean()

# Display the result
print(avg_salary_by_dept)
```

```python
# Filter employees with age > 30
filtered_df = df[df['Age'] > 30]

# Print the filtered DataFrame
# print(filtered_df)
# Select only 'name' and 'age' columns
result = filtered_df[['Name', 'Age']]

# Display the result
print(result)
```

```python
import numpy as np
```

```python
arr = np.linspace(1, 500, 100)
print(arr)
```

```python
mean = np.mean(arr)
median = np.median(arr)
std_dev = np.std(arr)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

```python
min_value = np.min(arr)
max_value = np.max(arr)

print("Minimum Value:", min_value)
print("Maximum Value:", max_value)
```

```python
import pandas as pd
df1=pd.read_csv("sales_data.csv")
print(df1)
```

```python
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime objects
df1['Date'] = pd.to_datetime(df1['Date'])

# Sort the DataFrame by date
df1 = df1.sort_values(by='Date')

# Plot the revenue trend over time
plt.figure(figsize=(10, 6))
plt.plot(df1['Date'], df1['Revenue'])
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.title('Revenue Trend Over Time')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Load the built-in tips dataset
tips = sns.load_dataset("tips")

# Display the first few rows
print(tips.head())
```

```python
# Create a scatter plot with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Relationship between Total Bill and Tip with Regression Line')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
df2=pd.read_csv("sales_data.csv")
print(df2)
```

```python
# Sort the DataFrame by 'Revenue' in descending order
df2_sorted_by_revenue = df1.sort_values(by='Revenue', ascending=False)

# Display the top 3 rows
print(df2_sorted_by_revenue.head(3))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create and save the figure
plt.figure(figsize=(10, 6))
sns.barplot(x='Product', y='Revenue', data=df2)
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue per Product')
plt.tight_layout()
plt.savefig('revenue_plot.png', dpi=300, bbox_inches='tight')  # Save with high DPI
plt.show()
```

```python
# Calculate the sum of revenue for each product
revenue_by_product = df2.groupby('Product')['Revenue'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(revenue_by_product, labels=revenue_by_product.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Revenue by Product')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig('revenue_pie_chart.png', dpi=400, bbox_inches='tight')
plt.show()
```
